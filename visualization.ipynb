```python
data= [1,2,3,4,6,8]
fig,ax =plt.subplots(figsize=(6,4))
ax.set_xlabel('no of elements', fontdict={'fontsize':10, 'color': 'red'})
ax.set_ylabel('Y - value')
ax.grid(True)
ax.set_title("Basics of Visualization", loc='right')
ax.plot(data, marker= 'o', markersize=10,
        markeredgecolor='black', markerfacecolor='navy')
ax.plot(data)
```

```python
data= [1,2,3,4,6,8,10,12]
data2=[22,23,30,36,55,7,7,11,12]
fig,ax =plt.subplots(figsize=(6,4))
ax.plot(data, marker= 'o', markersize=10,
        markeredgecolor='black', markerfacecolor='navy')
ax.plot(data2,marker='X',markersize=10,)
```

```python
data= [1,2,3,4,6,8,10,12]
data2=[22,23,30,36,55,7,11,13]
fig,ax =plt.subplots(figsize=(6,4))
ax.plot(data, marker= 'X', markersize=10,
        markeredgecolor='black', markerfacecolor='navy',
        linestyle='--',color='brown')
ax.plot(data2,marker='X',markersize=10,
        markeredgecolor='black', markerfacecolor='red',
        linestyle='--',color='green',linewidth=2,)
```

```python
data= [1,2,3,4,6,8,10,12]
data2=[22,23,30,36,55,7,11,13]
fig,ax =plt.subplots(figsize=(6,4))
ax.plot(data, marker= 'X', markersize=10,
        markeredgecolor='black', markerfacecolor='navy',
        linestyle='--',color='brown')
ax.plot(data2,marker='X',markersize=10,
        markeredgecolor='black', markerfacecolor='red',
        linestyle='--',color='green',linewidth=2,alpha=0.5)
```

```python
data= [1,2,3,4,6,8,10,12]
data2=[22,23,30,36,55,7,11,13]
fig,(ax1,ax2) =plt.subplots(figsize=(12,4),ncols=2)
ax1.plot(data, marker= 'X', markersize=10,
        markeredgecolor='green', markerfacecolor='navy',
        linestyle='--',color='brown')
ax2.plot(data2,marker='X',markersize=10,
        markeredgecolor='black', markerfacecolor='red',
        linestyle='--',color='green',linewidth=2,)
```
